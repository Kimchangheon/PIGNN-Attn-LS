```
import re
from pathlib import Path

def simplify_authors(bib_path, output_path=None):
    bib_text = Path(bib_path).read_text(encoding="utf-8")

    def simplify_author_field(match):
        author_field = match.group(0)
        authors_raw = match.group(1)
        # Split authors by " and " while avoiding the case where 'and' appears in names
        authors = [a.strip() for a in re.split(r'\s+and\s+', authors_raw) if a.strip()]
        if len(authors) > 2:
            # Extract first author’s last name and first name
            first_author = authors[0]
            simplified_authors = f"{first_author} and others"
        else:
            simplified_authors = " and ".join(authors)
        return f"author = {{{simplified_authors}}}"

    # Replace all author fields
    new_bib_text = re.sub(r"author\s*=\s*\{([^}]*)\}", simplify_author_field, bib_text)

    # Write to output (default: add '_simplified' suffix)
    if not output_path:
        output_path = bib_path.replace(".bib", "_simplified.bib")

    Path(output_path).write_text(new_bib_text, encoding="utf-8")
    print(f"✅ Simplified .bib file written to: {output_path}")

# Example usage:
# simplify_authors("references.bib")
```

```
simplify_authors("ref.bib")
```
